### 장르

```python
for i in range(0,498):
    link='https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='+'영화 '+df['title'][i]+' 정보'
    print(link)
    src=requests.get(link)
    cont=src.content.strip()
    html=bsp(cont,'html.parser')
    
    mgenre=html.find_all('div', 'info_group')
    if mgenre:
        genre=[html.find_all('div', 'info_group')[n].dd for n in range(2,3)]
        if genre[0]:
            genre_type=genre[0].text
            df.genre.iloc[i]=genre_type
        else:
            df.genre.fillna("")
    else:
        df.genre.fillna("")
```

### 평점, 시놉시스

```python
for i in range(0,498):
    stlink='https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='+'영화 '+df['title'][i]
    print(stlink)
    src=requests.get(stlink)
    cont=src.content.strip()
    html=bsp(cont,'html.parser')
    
    mstar=html.find_all('div', 'info_group')
    if mstar:
        star=html.find_all('div', 'info_group')[2].dd
        if star:
            star_type=star.text
            print(star_type)
            df.star.iloc[i]=star_type
        else:
            df.star.fillna("")
    else:
        df.star.fillna("")
        
    msys=html.find_all('div', 'text_expand _ellipsis')
    if msys:
        sys=[html.find_all('div', 'text_expand _ellipsis')[n].span for n in range(len(msys))]
        if sys:
            sys_type=sys[0].text
            print(sys_type)
            df.synopsis.iloc[i]=sys_type
        else:
            df.synopsis.fillna("")
    else:
        df.synopsis.fillna("")

    
    
```

### 감독, 배우

```python
for i in range(0,498):
    dlink='https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='+'영화 '+df['title'][i]
    print(dlink)
    src=requests.get(dlink)
    cont=src.content.strip()
    html=bsp(cont,'html.parser')
 
    mdir=html.find_all('div', 'title_box')
    if mdir:
        dir=[html.find_all('div', 'title_box')[n].strong.string for n in range(len(mdir))]
        if dir[0]:
            print(dir[0])
            df.director.iloc[i]=dir[0]
        if dir[1:5]:
            print(dir[1:5])
            act=dir[1:5]
            df.actor.iloc[i]=act
        else:
            df.actor.fillna("")
        
    else:
        df.actor.fillna("")
        
    
```
