```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
from selenium.webdriver import ActionChains
```

```python
driver = webdriver.Chrome()
url = "https://www.card-gorilla.com/team/detail/5"
driver.get(url)

benefit_list = []
card_list = []
card_details = []
action = ActionChains(driver)
#더보기 element 찾은후 다눌러서 없어질 때까지
while True:     
    try:
        some_tag = driver.find_element_by_css_selector('#q-app > section > div.team_detail > section > div > article.con_area > div > div.player_lst.results_lst > a')
        action.move_to_element(some_tag).perform()
        some_tag.click()
    except:    
        break
        time.sleep(5)

#카드 url 가져오기
a =driver.find_elements_by_css_selector(".b_view")
for i in a: 
    print(i.get_attribute("href"))
    card_details.append(i.get_attribute("href"))

#url 리스트를 통해 하나씩 정보 가져오기
for i in card_details:    
    #카드 자세히보기 이동    
    driver.get(i)
    time.sleep(3)    

    #카드 이름 가져오기
    card_names = driver.find_elements_by_css_selector(".card")
    for card_name in card_names:
        if card_name.text =="":
            continue
        else:
            print(card_name.text)
            card_list.append(card_name.text)

    # 혜택 가져오기                         
    try:
        benefits=[]
        for a in range(1,20):
            benefit = driver.find_element_by_css_selector("#q-app > section > div.card_detail.fr-view > section > div > article.cmd_con.benefit > div.lst.inner.faq_area > dl:nth-child({0}) > dt > i".format(a)).text
            # 유의사항 꼭 확인 나올시 benefit -> benefit list에 집어넣고 오류 발생으로 넘어가기
            if "꼭 확인" in benefit :
                benefit_list.append(benefits)
            else:
                print(benefit)
                benefits.append(benefit)

    except NoSuchElementException as e:
        pass
```
